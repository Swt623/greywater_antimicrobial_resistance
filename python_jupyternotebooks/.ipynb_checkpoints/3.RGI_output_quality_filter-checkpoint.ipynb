### Calculate RPKM from total reads, reference length and number of reads mapped to each allele
'ARO Term', 'All Mapped Reads'+sampleName, 'Reference Length', 'AMR Gene Family', 'Drug Class', 'Resistance Mechanism'

RPKM = numReads / (geneLength/1000 * totalNumReads/ 1000000 )

```python
# loose
mapq_th = 30
coverage_th = 50
len_cov = 75
```

```python
# strict
mapq_th = 50
coverage_th = 90
len_cov = 135
```

```python
# medium
mapq_th = 40
coverage_th = 80
len_cov = 120
```

```python
terms = ['ARO Term', 'AMR Gene Family', 'Drug Class', 'Resistance Mechanism', 'Resistomes & Variants: Observed Pathogen(s)','ARO Accession',
         'Reference Allele Source','Resistomes & Variants: Observed in Genome(s)','Resistomes & Variants: Observed in Plasmid(s)']
#rpkm_data = None
read_count = None
for file in files:
    sample = file[9:-24]
    df = pd.read_csv(file, delimiter = "\t",index_col=0)
    a = len(df['ARO Term'].unique())
    # filter by MAPQ, percentage coverage
    df = df[df["Average MAPQ (Completely Mapped Reads)"]>= mapq_th]
    df = df[(df["Percent Coverage"]>= coverage_th) | (df['Length Coverage (bp)']>=len_cov) ]
    #df = df[df["All Mapped Reads"]> read_th]
    print(f"{sample}")
    b = len(df['ARO Term'].unique())
    print(b)
    print(f"{b/a*100}%")
    #rpkm = df["All Mapped Reads"]/(df["Reference Length"]/1000 * reads[sample]/1000000)
    #rpkm = rpkm.to_frame().rename(columns={0: f"{sample}"})
    count = df["All Mapped Reads"].to_frame().rename(columns={"All Mapped Reads": f"{sample}"})
    if read_count is None:
        read_count = count 
        for term in terms: 
            read_count.loc[list(count.index), term] = df[term]
    else:
        read_count = pd.concat([read_count,count],axis=1).fillna(0)
        for term in terms:
            read_count.loc[list(count.index), term] = df[term]
    read_count = read_count.sort_index(axis = 1)
#rpkm_data.to_csv('ARGs/RGI/allele_rpkm_summary_strict.csv')
#ARO_data = rpkm_data.groupby(by="ARO Term").sum().drop(columns = ["ARO Accession"])
#ARO_data.to_csv('ARGs/RGI/ARO_rpkm_summary_strict.csv')
read_count.to_csv('ARGs/RGI/allele_read_count_strict.csv')
ARO_read = read_count.groupby(by="ARO Term").sum().drop(columns=['ARO Accession'])
ARO_read.to_csv('ARGs/RGI/ARO_read_count_strict.csv')
```

```python
DC_read = read_count.groupby(by="Drug Class").sum().drop(columns=['ARO Accession'])
DC_read.to_csv('ARGs/RGI/DrugClass_read_count_strict.csv')
```

```python
sample_id = list(ARO_data.columns)
```

```python
# loose
bitscore = 50
identity = 80.0
per_length = [50,110]
```

```python
# medium
bitscore = 50
identity = 90
per_length = [80,110]
```

```python
# strict
bitscore = 50
identity = 95.0
per_length = [90,110]
```

```python
for sample in sample_id:
    print(f"Processing sample {sample}.")
    RGI_out = f"ARGs/contig-RGI/metaspades/{sample}.txt"
    df1 = pd.read_csv(RGI_out, delimiter = "\t", index_col = 0)
    # get contig ID for RGI output file
    contig_list = ["_".join(name.split("_")[:-1]) for name in list(df1["Contig"])]
    x = len(df1["Best_Hit_ARO"].unique())
    print(f"AMR genes found in {len(set(contig_list))} contigs.")
    print(f"There are {x} AMR gene detected.")
    
    df1["Contig_ID"] = contig_list
    # filter 
    contig_ARG = df1.set_index("Contig_ID")
    contig_ARG = contig_ARG[(contig_ARG["Best_Identities"] >= identity) & (contig_ARG["Best_Hit_Bitscore"] >= bitscore)]
    contig_ARG = contig_ARG[(contig_ARG["Percentage Length of Reference Sequence"] >=  per_length[0]) & (contig_ARG["Percentage Length of Reference Sequence"] <=  per_length[1])]
    y = len(contig_ARG["Best_Hit_ARO"].unique())
    q = y/x*100
    print(f"There are {y} ({q}%) AMR gene passed QC. \n")
    contig_ARG = contig_ARG.rename(columns={"Best_Hit_ARO": "ARO Term"})
    contig_ARG.to_csv(f'ARGs/contig-RGI/metaspades/{sample}_filter_strict.csv')
```

```python
ARO_data = pd.read_csv('ARGs/RGI/ARO_rpkm_summary_strict.csv', index_col = 0)
bwt = []
main = []
common = []
cum_rpkm = []
sum_counts = []
for sample in sample_id: 
    # RGI bwt output
    ARO_bwt = list(ARO_data[ARO_data[f'{sample}'] != 0.0].index)
    n = len(ARO_bwt)
    bwt.append(n)
    x = ARO_data[f'{sample}'].sum()
    cum_rpkm.append(x)
    # RGI main output
    ARO_main_data = pd.read_csv(f'ARGs/contig-RGI/metaspades/{sample}_filter_strict.csv', index_col = 0)
    ARO_main = list(ARO_main_data['ARO Term'].unique())
    m = len(ARO_main)
    main.append(m)
    y = ARO_main_data.shape[0]
    sum_counts.append(y)
    overlap = [arg for arg in ARO_bwt if arg in ARO_main]
    o = len(overlap)
    common.append(o)
    print(f"{sample} has {o} overlapping ARGs ({o/n*100}%, {o/m*100}%)")
ARG_quant = pd.DataFrame(index = sample_id)
ARG_quant["short reads"] = bwt
ARG_quant["assembly"] = main
ARG_quant["common"] = common
ARG_quant["short reads only"] = ARG_quant["short reads"]-ARG_quant["common"]
ARG_quant["assembly only"] = ARG_quant["assembly"]-ARG_quant["common"]
ARG_quant["short reads cumulative rpkm"] = cum_rpkm
ARG_quant["assembly ARG counts"] = sum_counts
ARG_quant_strict = ARG_quant
```

```python
terms = ['AMR Gene Family', 'Drug Class', 'Resistance Mechanism']
count_data = None
for sample in sample_id:
    df = pd.read_csv(f'ARGs/contig-RGI/metaspades/{sample}_filter_strict.csv', index_col=1)
    df1 = df.groupby(by = 'ARO Term').first()
    df['ARG count'] = 1
    df2 = df.groupby(by = 'ARO Term').sum()
    #print(df2.head(2))
    count = df2['ARG count'].to_frame().rename(columns={"ARG count": f"{sample}"})
    #print(df[terms].head(2))
    if count_data is None:
        count_data = count
        for term in terms: 
            count_data.loc[list(count.index), term] = df1[term]
    else:
        count_data = pd.concat([count_data,count],axis=1).fillna(0)
        for term in terms:
            count_data.loc[list(count.index), term] = df1[term]
    count_data = count_data.sort_index(axis = 1)
count_data.to_csv('ARGs/contig-RGI/metaspades/ARO_count_summary_strict.csv')
```
