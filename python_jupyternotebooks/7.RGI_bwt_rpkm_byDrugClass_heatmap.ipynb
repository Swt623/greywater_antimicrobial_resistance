```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
import pylab
```

```python
file = "RGI/all_sample_rpkm_DrugClass_strict_ungrouped.csv"
DrugClass_df = pd.read_csv(file,delimiter = ",")
DrugClass_df.head()
```

```python
DrugClass_single = DrugClass_df.groupby(["Drug Class"]).sum()
```

```python
DrugClass_single.head(10)
```

```python
DrugClass_single.to_csv("RGI/DrugClass_rpkm_strict_summary.csv")
```

```python
file = "RGI/DrugClass_rpkm_strict_withBiomassRatio.csv"
DrugClass_BiomassRatio_corrected_df = pd.read_csv(file,delimiter = ",",index_col=0)
DrugClass_BiomassRatio_corrected_df.head()
```

```python
DrugClass_BiomassRatio_corrected_df['Sum'] = DrugClass_BiomassRatio_corrected_df.sum(axis=1)
```

```python
DrugClass_BiomassRatio_corrected_df = DrugClass_BiomassRatio_corrected_df.sort_values(by=['Sum'],ascending=False).drop(columns=['Sum'])
```

```python
plt.rcParams['figure.figsize'] = [8, 8]
colormap = sns.color_palette("Blues")
sns.heatmap(DrugClass_BiomassRatio_corrected_df, cmap=colormap, cbar_kws={'label': 'ARG abundance corrected with biomass change factor (RPKM)'})
plt.xticks(rotation=30, ha="right")
plt.savefig('figures/Heatmap-RGI_bwt_DrugClass_RPKM_strict_with_biomass_change_factor-correct.pdf',bbox_inches='tight')
```
