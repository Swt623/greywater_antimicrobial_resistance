## functions to clean index names

```python
# rename index names
def clean_index_names(df, sep): # read in a dataframe and a separator for the index name cleaning
    # get the list of index names
    index_list = list(df.index)
    for name in index_list:
        name_list = name.split(sep)
        n = len(name_list)
    # get the last nonempty item of the name list 
        for i in range (1,n):
            if name_list[-i] != "":
                df.rename(index = {name: name_list[-i]}, inplace=True)
                break
    return df
```

## Metaxa2 taxonomy output cleaning

### Get mapped read numbers for each sample

```python
# read in metaxa2 reads numbers from file, each line is [sample ID]: [reads]
reads = "taxonomy/reads_summary.txt"
# construct a dictionary of sample:reads.
reads_dict = {}
with open(reads, "r") as f: 
    for line in f.readlines():
        col = line.split(":")
        reads_dict[col[0]] = int(col[1])
```
